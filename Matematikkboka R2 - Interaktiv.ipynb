### 0.2 Kalender
<a id='1.1-Kalender'></a>

<a href="#Innholdsfortegnelse">⬆ Tilbake til innholdsfortegnelse</a>

```python
import calendar
from datetime import date

def display_calendar():
    year = int(input("Skriv år: "))
    month = int(input("Skriv måned (1-12): "))

    cal = calendar.TextCalendar(calendar.MONDAY)
    
    # Print the month calendar
    month_calendar = cal.formatmonth(year, month)
    print(month_calendar)

    # Print week numbers
    print("Ukenumre:")
    month_weeks = cal.monthdayscalendar(year, month)
    for week in month_weeks:
        # Find the first day of the week that is not 0 (which means no day)
        first_day = next((day for day in week if day != 0), None)
        if first_day:
            week_number = date(year, month, first_day).isocalendar()[1]
            print(f"Uke {week_number}: {week}")

display_calendar()
```

## 1.1 Tallfolger

<a href="#Innholdsfortegnelse">⬆ Tilbake til innholdsfortegnelse</a>

```python
# Tallføger og rekursive følger
import sympy as sp

def rekursiv_r2_loser_v2():
    print("\n=== R2 Rekursiv Følge Løser (Fikset) ===")
    print("Her skriver du uttrykket for neste ledd ved hjelp av det forrige.")
    print("Bruk bokstaven 'a' for det forrige leddet (a_n-1).")
    
    # Definer symbolet 'a' som representerer a_{n-1}
    # real=True hjelper sympy å skjønne at dette er et tall
    a = sp.Symbol('a', real=True)
    n = sp.Symbol('n', integer=True) # Legger til n i tilfelle du trenger det (f.eks a_n = a + n)
    
    while True:
        print("\n-----------------------------------")
        raw_input = input("Formel (f.eks '3*a' eller '0.5*a + 4'):\n> a_n = ").strip()
        
        if raw_input.lower() in ['q', 'quit', 'exit']:
            break
            
        try:
            # --- VIKTIG ENDRING HER ---
            # Vi forteller sympify at bokstaven 'a' i teksten er VARIALBELEN a vi lagde over.
            # Dette kobler teksten din til matematikken.
            uttrykk = sp.sympify(raw_input, locals={'a': a, 'n': n})
            
            a1_inp = input("Startverdi a1 = ")
            # Håndterer brøk i startverdi (f.eks '1/2')
            a1_val = float(sp.sympify(a1_inp)) 
            
            antall = int(input("Antall ledd å beregne (f.eks 6): ") or 6)
            
            # --- STEG 1: Beregn leddene ---
            ledd = [a1_val]
            print(f"\n[1] Utregning av de første {antall} leddene:")
            print(f"  n=1: {a1_val}")
            
            for i in range(2, antall + 1):
                # Sett inn forrige ledd for 'a', og nåværende indeks for 'n' (hvis formelen bruker n)
                # Vi bruker float() rundt ledd[-1] for å hindre at brøker vokser seg enorme i minnet
                neste_verdi = uttrykk.subs({a: ledd[-1], n: i})
                
                # Sikkerhetskonvertering til flyttall for plotting/visning
                tall_verdi = float(neste_verdi)
                ledd.append(tall_verdi)
                
                print(f"  n={i}: {tall_verdi:.2f}")
                
            # --- STEG 2: Finn grenseverdi (Fikspunkt) ---
            print("\n[2] Analyse av grenseverdi:")
            
            # Vi løser ligningen: x = f(x) (dvs a = uttrykk)
            ligning = sp.Eq(a, uttrykk.subs(n, sp.Symbol('n'))) # Fjerner n hvis den er der for grensen
            
            try:
                kandidater = sp.solve(ligning, a)
            except:
                kandidater = []

            if not kandidater:
                print("  Ingen åpenbar grenseverdi (ligningen a = f(a) har ingen løsning).")
            else:
                print(f"  Hvis følgen konvergerer, må den gå mot en av disse:")
                for k in kandidater:
                    try:
                        print(f"  L = {k} (≈ {float(k):.2f})")
                    except:
                        print(f"  L = {k}")
                
                # Sjekk faktisk utvikling
                siste = ledd[-1]
                nest_siste = ledd[-2]
                diff = abs(siste - nest_siste)
                
                print("\n[3] Konklusjon:")
                if diff > 1000: 
                    print("  Følgen DIVERGERER tydelig (går mot uendelig).")
                elif diff < 1e-5:
                    print(f"  Følgen ser ut til å ha KONVERGERT mot ≈ {siste:.2f}")
                else:
                    print("  Følgen endrer seg fortsatt. Beregn flere ledd for å være sikker.")

        except Exception as e:
            print(f"⚠️  Feil oppstod: {e}")
            print("Tips: Sjekk at formelen er gyldig matte. Bruk punktum for desimaltall (0.5).")

if __name__ == "__main__":
    rekursiv_r2_loser_v2()
```

```python
# Konvergens og geometriske rekker
import sympy as sp

def r2_tallfolge_analyse():
    print("\n=== Sinus R2 Tallfølge-analyser ===")
    print("Her bruker vi symbolsk matematikk for å finne eksakte svar.")
    
    # Definer n som et symbol
    n = sp.Symbol('n')
    
    while True:
        print("\n-----------------------------------")
        try:
            inp = input("Skriv inn formel for a_n (eller 'q' for å avslutte)\nF.eks: 3 + (1/2)**n  eller  (2*n + 1)/(n + 5)\n> ").strip()
            
            if inp.lower() == 'q':
                break

            # 1. Tolk uttrykket
            # Vi konverterer tekstinput til et matematisk uttrykk
            uttrykk = sp.sympify(inp)
            
            # 2. Be om startindeks
            start_n = int(input("Hvilken n starter vi på? (f.eks. 1): ") or 1)

            # --- DEL A: De neste 5 leddene ---
            print(f"\n[1] De 5 leddene fra n={start_n}:")
            verdier = []
            for i in range(start_n, start_n + 5):
                verdi = uttrykk.subs(n, i)
                verdier.append(verdi)
                # Vi printer både eksakt (brøk) og desimaltall
                print(f"  a_{i} = {verdi}  \t(≈ {float(verdi):.5f})")

            # --- DEL B: Sjekk for Geometrisk følge (R2 pensum) ---
            # En følge er geometrisk hvis a_{n+1} / a_n er en konstant k
            a_n = uttrykk
            a_n_plus_1 = uttrykk.subs(n, n + 1)
            kvotient = sp.simplify(a_n_plus_1 / a_n)
            
            print("\n[2] Type følge:")
            if kvotient.is_constant():
                k = float(kvotient)
                print(f"  Dette er en GEOMETRISK følge med k = {k}")
                if abs(k) < 1:
                    print("  Siden |k| < 1, vil den uendelige rekken KONVERGERE.")
                    # Beregn summen S = a1 / (1-k)
                    a1 = uttrykk.subs(n, start_n)
                    S = a1 / (1 - kvotient)
                    print(f"  Sum av uendelig rekke S = {S} (≈ {float(S):.4f})")
                else:
                    print("  Siden |k| >= 1, vil rekken DIVERGERE (summen går mot uendelig).")
            else:
                print("  Dette er ikke en enkel geometrisk følge (kvotienten varierer med n).")
                print("  Vi bruker generell grenseverdi-analyse.")

            # --- DEL C: Grenseverdi (Limes) ---
            print("\n[3] Grenseverdi når n -> ∞:")
            grense = sp.limit(uttrykk, n, sp.oo)
            
            if grense == sp.oo or grense == -sp.oo:
                print(f"  Følgen DIVERGERER (går mot {grense}).")
            else:
                print(f"  Følgen KONVERGERER.")
                print(f"  lim(a_n) = {grense}")

        except Exception as e:
            print(f"⚠️  Noe gikk galt: {e}")
            print("Tips: Bruk ** for opphøyd (f.eks 2**n), og pass på parenteser.")

if __name__ == "__main__":
    r2_tallfolge_analyse()
```
