# $\color{black}{\text{Kapittel 1 - Grunnlaget}}$

```python
# UTFORSK ANDREGRADSFORMELEN
from math import sqrt

print("Vi skal løse andregradslikningen ax^2+bx+c=0")

a = float(input("a ="))

if a == 0:
  print("Med a = 0 blir det ikke en andregradslikning.")
else:
  b = float(input("b ="))
  c = float(input("c ="))
  d = b**2-4*a*c
  
  if d < 0:
    print("Ingen løsning")
  elif d == 0:
    x1 = -b/(2*a)
    print("En løsning x =", round(x1, 2))
  else:
      x1 = (-b+sqrt(d))/(2*a)
      x2 = (-b-sqrt(d))/(2*a)
      print("To løsninger x =", round(x1,2), "og x =", round(x2,2))
```

# $\color{lightblue}{\text{Kapittel 5 - Eksponentialfunksjoner}}$

```python
# Eksempel s. 199 - 200
from math import exp

def q(p):
  return 1000*exp(-0.03*p)

def I(p):
  return p*q(p)

a = 10
b = 40
p = a
pmax = p
Imax = I(p)

for p in range(a, b+1):
  if I(p) > Imax:
    pmax = p
    Imax = I(p)

print("Inntekten er størst når prisen er",pmax,"kr.")
print("Etterspørselen er da",round(q(pmax),0),"enheter per uke.")
print("Inntekten er",round(Imax,0),"kr per uke.")
```

# $\color{coral}{\text{Kapittel 6 - Sannsynlighet}}$

```python
# Binominalkoeffisienten s. 217
def fakultet(n):
  fak = 1
  for i in range(1, n+1):
    fak *= i
  return fak

def nCr(n, k):
  return fakultet(n)/(fakultet(k)*fakultet(n-k))

print(nCr(7, 5))
```

```python
# Permutasjoner s. 225 Dette er et Pythonprogram som kan fortelle hvor mange ordnede utvalg uten tilbakelegging det finnes.
def fakultet(n):
  fak = 1
  for i in range(1, n+1):
    fak *= i
  return fak

def nPr(n, k):
  return fakultet(n)/fakultet(n-k)

print(nPr(10, 7))
```

```python
# Simulering høyrevelgere s. 234
# Dette er et Pythonprogram som tar i bruk randomgenerator til å simulere en meningsmåling. 
# Dette kan sammenlignes med det faktiske valgresultatet også kan det vurderes hvor pålitelig meningsmålingen er.
from random import randint

høyrevelgere = 0
antallintervju = 1200

for intervju in range(antallintervju):
  if randint(1,100) <= 25:
    høyrevelgere += 1

andel = høyrevelgere/antallintervju

print("Andel høyrevelgere er", round(andel*100, 1), "%")
```

```python
# Simulering terningkast s. 235
# Dette er et Pythonprogram du kan bruke til å undersøke hvor mange kast med en terning vi må bruke for å få en sekser.
from random import randint

antallkast = 0
terning = 1

while terning < 6:
  terning = randint(1,6)
  antallkast += 1

print("Antall kast =", antallkast)
```

```python
# Dette er et Pythonprogram som kan brukes til å løse eksempelet på s. 270 - 271. 
def fakultet(n):
  fak = 1
  for i in range(1, n+1):
    fak *= i
  return fak

def nCr(n,k):
  return fakultet(n)/(fakultet(k)*fakultet(n-k))

def hypergeometrisk(n,n1,k,a):
  return nCr(n1,a) * nCr(n-n1,k-a) / nCr(n,k)

n = 30    #Antallet vi trekker fra
n1 = 18   #Antallet av type 1
k = 10    #Antallet vi trekker 

print("P(X = 6) =", round(hypergeometrisk(n,n1,k,6),4))
```

```python
# Eksempel s. 263-264
def fakultet(n):
  fak = 1
  for i in range(1, n+1):
    fak *= i
  return fak

def nCr(n,k):
  return fakultet(n)/(fakultet(k)*fakultet(n-k))

def binomisk(n,p,k):
  return nCr(n,k) * p**k * (1-p)**(n-k)

n = 30
p = 0.15

print("P(X = 5) =", round(binomisk(n,p,5),4))
```
