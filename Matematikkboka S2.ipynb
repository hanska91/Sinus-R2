```python
# Serielån, s.29
def restlån(n):
  return lån - (n - 1)*avdrag

def terminbeløp(n):
  return avdrag + månedsrentefot/100 * restlån(n)

lån = float(input("Lån i kroner:"))
antall_år = int(input("Antall år i nedbetalingstid:"))
rentefot = float(input("Rentefot:"))

n = antall_år * 12            #n er antall terminer
avdrag = lån/n
månedsrentefot = rentefot/12

sum = (terminbeløp(1)+terminbeløp(n))/2 * n
rente = sum - lån

print(f"Han betaler til sammen {round(sum)} kr.")
print(f"Av dette er {round(rente)} kr rente")
```

```python
# Annuitetslån, s 49
lån = float(input("Lånebeløp:"))
rentefot = float(input("Rentefot:"))
antall_år = int(input("Antall år i nedbetalingstid:"))
terminer_per_år = int(input("Antall terminer per år:"))

n = antall_år * terminer_per_år    # n er antall terminer totalt
v = ((1 + rentefot/100)**(1/terminer_per_år))
F = (1 - 1/v**n)/(v - 1)
terminbeløp = lån/F

print(f"Terminbeløpene er på {round(terminbeløp)} kr.")
print(f"Rentene var til sammen {round(n*terminbeløp-lån)} kr.")
```

# $\color{green}{\text{Kapittel 3 - Integrasjon}}$

```python
# Eksempel side 122
from math import sqrt

def f(x):
  return 3*sqrt(x)

a = float(input("Nedre grense?"))
b = float(input("Øvre grense?"))
n = int(input("Antall rektangler?"))

delta_x = (b-a)/n
sum = 0

for i in range(1, n+1):
  sum += f(a+(i-1)*delta_x)*delta_x

print(f"{n} rektangler gir tilnærmingsverdien {sum:.3f}.")
```

```python
# Binomisk sannsynlighetsmodell, s.195
from scipy.stats import binom

# P(x) = P(X = x)
def P(x):
  return binom.pmf(x, n, p)

# Pml(x) = P(X < x) (ml: mindre enn eller lik)
def Pml(x):
  sum = 0
  for i in range(x+1):
    sum += P(i)
  return sum

n = 10
p = 0.25
print("x | P(X = x)")
for x in range(0, n+1):
  print(f"{x} | {P(x):.3f}")

print()
print(f"P(X < 3) = {Pml(2):.3f}")
```

```python
# Eksempel standard normalfordeling, s.242
from scipy.stats import norm

def Fi(z):
  return norm.cdf(z)
  
def Pml(x):
  return Fi((x-my)/sigma)

my = 180
sigma = 7

print(f"a) P(X < 190) = {Pml(190):.4f}")
print(f"b) P(173 < X < 185) = {Pml(185)-Pml(173):.4f}")
print(f"c) P(X > 195) = {1 - Pml(195):.4f}")
```

```python
# I Utforsk meningsmåling på side 244-246 i boka simulerer vi en meningsmåling. 
# Her finner du programmene vi bruker.
from pylab import figure, show
from random import randint

def prosent(x):
  return x/antall_intervju * 100

antall_intervju = 100
antall_serier = 1000
maks = 0
min = antall_intervju

#Lager ei liste med 0
fordeling = [0]
for i in range(antall_intervju+1):
  fordeling += [0]
  
for serie in range(antall_serier):
  sum = 0
  for intervju in range(antall_intervju):
    if randint(1, 1000) <= 135:
      sum += 1
  fordeling[sum] += 1
  if sum > maks:
    maks = sum
  if sum < min:
    min = sum

#Tilpasser koordinatsystemet
x = []
y = []
for i in range(min, maks+1):
  x += [prosent(i)]
  y += [fordeling[i]]

#Lager histogrammet
histo = figure()
akser = histo.add_subplot()
akser.set_xlabel("Prosent")
akser.set_ylabel("Antall intervju")
akser.bar(x,y)
show()
```
