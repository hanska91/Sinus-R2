# Eksamen R2 vår 2024: Oppgave 1 - Del 2

Denne oppgaven fra del 2 handler om å analysere banen til et hjørnespark. Vi skal løse oppgaven numerisk med Numpy/Scipy, symbolsk med Sympy og skissere den numeriske løsningen i tre dimensjoner med Matplotlib. Vi skal bruke oppgaven til å undersøke hvordan vi kan bruke datetime og Babel til å regne med og presentere tidspunkter.

Oppgavetekst
En fotballspiller skal ta et hjørnespark (corner) på en fotballbane. Posisjonen r til ballen etter t sekunder er gitt ved

r(t) = [ 30t, 5t, 7t – 4,9t² ]

Her er posisjonen gitt i meter, og koordinatsystemet er lagt slik at origo er i hjørnet av fotballbanen, x-aksen går langs kortsiden og y-aksen går langs langsiden.

a) Hvor stor er farten til ballen idet den blir skutt?

b) Hvor langt fra hjørnemerket er ballen når den treffer fotballbanen igjen?

c) Hvor stor er farten til ballen når den er på sitt høyeste? Hvor høyt over fotballbanen er ballen da?



Numerisk løsning og grafisk framstilling med Numpy

```python
# Begynner med å legge inn og tegne vektorfunksjonen i tre dimensjoner med Matplotlib.
# Lager en liste med t-verdier fra ballen blir sparket til ballen treffer bakken.
# Importerer nødvendige biblioteker.
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import sympy as sp

#Importerer nødvendige funksjoner
from scipy.optimize import minimize_scalar, root_scalar
from sympy import Interval, oo, Point, Rational, solveset, Symbol

# Justerer grafikkingstillingene
plt.rcParams['figure.figsize'] = 7, 7

# Numerisk løsning og grafisk framstilling med Numpy
# Grafisk framstilling
# Begynner med å legge inn og tegne vektorfunksjonen i tre dimensjoner med Matplotlib.
# Lager en liste med t-verdier fra ballen blir sparket til ballen treffer bakken.
tmax = 7/4.9
T = np.linspace(0, tmax)

# Lager funksjoner for x-, y-, og z-koordinatene til ballen.
def xun(t):
    return 30*t
 
def yun(t):
    return 5*t
 
def zun(t):
    return 7*t - 4.9*t**2

# Finner x-, y- og z-verdiene.
X = xun(T)
Y = yun(T)
Z = zun(T)

# Har nok informasjon til å skissere banen til ballen. Vi tegner også «skyggen» til banen på bakken.
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(X, Y, Z)
ax.plot(X, Y, color="tab:blue", linestyle="--")
ax.plot([0, 50], [0, 0], [0, 0], color="black")
ax.plot([0, 0], [0, 10], [0, 0], color="black")
plt.show()
```

```python
# a) Startfartene er lengden av hastighetsvektoren r‘. 
# Siden funksjonen ikke eksisterer for t < 0 bruker vi foroverdifferansen til å finne hastighetsvektoren 
# som er den deriverte til posisjonsvektoren. Bruker diskretiseringen fra den grafiske framstillingen.
dt = T[1] - T[0]
print(dt)
```

```python
R = np.array([X, Y, Z]).T
dR = (R[1:] - R[:-1])/dt
va = dR[0]
print(va)
```

```python
# Lengden til en vektor er kvadratroten av vektorens prikkprodukt med seg selv.
starthastighet = np.sqrt(va.dot(va))
print(starthastighet)
```

```python
print(f"1a) Starthastigheten er {starthastighet:.3g} m/s")
```

```python
# b) Landingspunkt
# Løser for z-komponenten: Z(t) = 0 med root_scalar for å finne tiden.

resultat = root_scalar(zun, x0=1)
print(resultat)
```

```python
# Bruker tiden til å finne avstandsvektoren.
tb = resultat.root
rb = np.array([xun(tb), yun(tb), zun(tb)])
print(rb)
```

```python
# Svaret blir lengden til avstandsvektoren.
avstand = np.sqrt(rb.dot(rb))
print(avstand)
```

```python
print(f"1b) Ballen er {avstand:.3g} m fra hjørnemerket når den treffer fotballbanen igjen.")
```

```python
# 1 c) Toppunkt
# Ballen er på sitt høyeste når z-komponenten er størst. Finner tiden og z-komponenten med minimize_scalar. 
# Bytter fortegnet på z-komponenten siden funksjonen vi bruker er en minimeringsfunksjon.

fun = lambda t: -zun(t)
resultat = minimize_scalar(fun)
print(resultat)
```

```python
# Henter ut tiden og høyden.
tc = resultat.x
zc = -resultat.fun
```

```python
# Finner hastighetsvektoren med sentraldifferansen.
vc = np.array([xun(tc + dt/2) - xun(tc - dt/2),
               yun(tc + dt/2) - yun(tc - dt/2),
               zun(tc + dt/2) - zun(tc - dt/2)])/dt
```

```python
# Farten er lengden på hastighetsvektoren.
fc = np.sqrt(vc.dot(vc))
print(f"1c) Farten til ballen er {fc:.3g} m/s når den er på sitt høyeste.")
print(f"1c) Ballen er da {zc:.3g} m over bakken.")
```

```python
# Tegner situasjonen. Regner først ut koordinatene i toppunktet

xc, yc = xun(tc), yun(tc)
rc = np.array([xc, yc, zc])

# Lager en tegnefunksjon som skisserer fotballbanen og ballbanen.

def tegn():
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.plot(X, Y, Z, label="r(t)", color="tab:blue")
    ax.plot(X, Y, color="tab:blue", linestyle="--")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    ax.azim, ax.elev = -30, 15
    fig.tight_layout()
    ax.plot([0, 50], [0, 0], [0, 0], color="black")
    ax.plot([0, 0], [0, 10], [0, 0], color="black")
    ax.elev = 15
    ax.azim = -30
    return fig, ax

# Tegner banen og legger til toppunktet og hastighetsvektor.

fig, ax = tegn()
ax.scatter(*rc, color="tab:orange", label="toppunkt")
ax.plot(*zip(rc, rc + vc), color="tab:orange")
ax.plot([xc, xc], [yc, yc], [0, zc], color="tab:grey")
ax.legend()
plt.show()
```

```python
# Animasjon
# Vi kan lage en 3D-animasjon som viser ballens posisjon og hastighetsvektor.

# Siden vi vil vise hastighetsvektoren må vi finne denne for alle verdier av T.

VX = (X[1:] - X[:-1])/dt
VY = (Y[1:] - Y[:-1])/dt
VZ = (Z[1:] - Z[:-1])/dt

# Vi finner også tallverdien til farten.

V = np.sqrt(VX**2 + VY**2 + VZ**2)

# Tegner først de faste elementene, før jeg legger til de variable elementene. 
# Deretter lager jeg en animasjonsfunksjon som oppdaterer de variable elementene. 
# Til slutt lager jeg et FuncAnimation-objekt som styrer animasjonen.

fig, ax = tegn()
posisjon, = ax.plot(X[0], Y[0], Z[0], marker="o", label="tid = 0", color="tab:orange")
hastighet, = ax.plot([0, VX[0]], [0, VY[0]], [0, VZ[0]],
                     label=f"hastighet = {V[0]:.1f}", color="tab:orange")
høyde, = ax.plot([X[0], X[0]], [Y[0], Y[0]], [0, VZ[0]],
                     label=f"høyde = {Z[0]:.1f}", color="tab:grey")
forklaring= ax.legend(loc=1)
 
def animate(i):
    #i % len(pdata.T) gjør at animasjonen går i loop.
    n = i % (len(T) - 1)
 
    # Henter ut posisjonsdata
    x, y, z = X[n], Y[n], Z[n]
    posisjon.set_data([x], [y]) # Oppdaterer x- og y- posisjon
    posisjon.set_3d_properties([z]) # Oppdaterer z-posisjon
 
 
    # Henter ut hastighetsvektoren.
    xv, yv, zv = VX[n], VY[n], VZ[n]
 
    # Legger inn x- og y-komponentene til start- og
    hastighet.set_data([x, x + xv], [y, y + yv])
 
    # Legger inn z-komponentene til start og- sluttpunkt
    # for hastighetsvektoren.
    hastighet.set_3d_properties([z, z + zv])
 
    # Legger til høyden
    høyde.set_data([x, x], [y, y])
    høyde.set_3d_properties([0, z])
 
    #Oppdaterer forklaringene
    t = T[n]
    lab1 = f"tid: {t:.2f} s"
    forklaring.get_texts()[1].set_text(lab1)
 
    v = V[n]
    lab2 = f"hastighet: {v:.1f} m/s"
    forklaring.get_texts()[2].set_text(lab2)
 
    lab3 = f"høyde: {z:.1f} m"
    forklaring.get_texts()[3].set_text(lab3)
 
    return posisjon, hastighet, høyde, forklaring
 
ani = animation.FuncAnimation(fig, animate, interval=100,
          blit=True, save_count=50, repeat=True)
 
ani.save("oppgave1.gif")
 
plt.show()
```

## Eksamen R2 vår 2024: Oppgave 1 - Del 2 Eksakt løsning med Sympy

```python
# a) Startfart
# Legger inn funksjonsuttrykkene til x, y og z-komponentene.

t = Symbol("t")
x = 30*t
y = 5*t
z = 7*t - Rational(49, 10)*t**2

# Legger inn posisjonsvektoren som et Point-objekt.

r = Point(x, y, z)

# Deriverer elementvis for å finne et uttrykk for hastighetsvektoren.

dr = Point([e.diff() for e in r])
print(dr)
```

```python
# Setter t = 0 inn i uttrykket elementvis.

dr0 = Point([e.subs(t, 0) for e in dr])
print(dr0)
```

```python
# Farten er lengden av hastighetsvektoren.

startfart = sp.sqrt(dr0.dot(dr0))
print(f"1a) Starthastigheten er {startfart} m/s = {startfart.evalf(5)} m/s")
```

```python
# b) Nedslagspunkt
# Løser for z-komponenten lik 0 med solveset.

løsning = solveset(z, t, Interval(0, oo))
print(løsning)
```

```python
# Det første tidspunktet er starten, nedslaget er den siste verdien.

tb = max(løsning)
print(tb)
```

```python
# Bruker tiden til å finne avstandsvektoren.

rb = Point([e.subs(t, tb) for e in r])
print(rb)
```

```python
# Svaret blir lengden til avstandsvektoren.

avstand = sp.sqrt(rb.dot(rb))
print(avstand)
```

```python
print(f"1b) Ballen er {avstand} m = {avstand.evalf(5)} m fra hjørnemerket når den treffer fotballbanen igjen.")
```

```python
# c) Toppunkt
# Ballen er på sitt høyeste når z-komponenten er størst. Finner tiden for dette ved å løse z‘(t) = 0 med solveset.

løsning = solveset(z.diff())
print(løsning)
```

```python
tc = list(løsning)[0]
zc = z.subs(t, tc)

# Finner toppunktet.

vc = Point([e.subs(t, tc) for e in dr])
fc = sp.sqrt(vc.dot(vc))
 
print(f"1c) Farten til ballen er {fc} m/s = {fc.evalf(5)} m/s når den er på sitt høyeste.")
```

```python
print(f"1c) Ballen er da {zc} m = {zc.evalf(5)} m over bakken.")
```

Skisse av omdreiingslegeme
Vi kan bruke randdataene vi har funnet til å skissere pæren i tre dimensjoner. 
Lager en funksjon som tar inn en liste med randen og skalafaktoren og bruker disse til å tegne et omdreiingslegeme.

```python
def tegn_omdreiingslegeme(R, d, farge="tab:green"):
    #Lager et figur- og et akseobjekt.
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
 
 
    #Vi lager variabler for parametriseringen av kjegle-objektet.
    #u er posisjon langs x-aksen
    xmin, xmax = 0, len(R)*d
    u = np.linspace(xmin, xmax, len(R))
 
    #v er rotasjonsvinkel om x-aksen.
    n = 50
    v = np.linspace(0, 2 * np.pi, n)
 
 
    #Lager datamatrisen.
    #X-matrisen har de samme x-koordinatene for hver rotasjonsvinkel.
    X = np.repeat([u], n, 0)
 
    #Y- og Z-matrisen får forskjellige verdier avhengig av rotasjonsvinkelen.
    Y = np.outer(np.sin(v), R)
    Z = np.outer(np.cos(v), R)
 
    #Vi tegner omdreiingslegemet
    ax.plot_surface(X, Y, Z, color=farge, alpha = 0.5)
 
    #Vi justerer og pynter på aksene.
    ax.plot([xmin - 2, xmax + 2], [0, 0], color="black")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    ax.axis('scaled')
    ax.elev = 15
    fig.set_size_inches(7, 7)
    fig.tight_layout()
    return fig, ax
 
fig, ax = tegn_omdreiingslegeme(R*dx, dx)
plt.show()
```

Denne oppgaven fra del 2 handler om å bruke en sinusfunksjon til å styre en lysbryter i løpet av et år. Vi skal løse oppgaven med Sympy.

Oppgavetekst
En sensor skal slå på utelyset foran ytterdøra til et hus. Lyset blir slått på T(x) timer etter midnatt. T(x) er gitt ved

T(x) = 4·sin(0,0055π·x – 0,5π) + 19.

x er antall dager etter 31. desember 2023 slik at x= 1 svarer til 1. januar 2024. Tidspunktet sensoren slår på utelyset, varierer fra kl. 15:00 til kl. 23:00, og det varierer periodisk i løpet av et år. Den 1. april slår lyset seg på kl. 19:00.

a) Forklar hvordan de ulike verdiene i modellen T(x) passer med opplysningene gitt ovenfor.

b) Når i 2024 vil tidspunktet da lyset slår seg på, flytte seg 3 minutter per dag?

c) Når endrer dette tidspunktet seg raskest, og hvor stor er endringen da?



```python
# Oppsett
# Importerer nødvendige biblioteker.

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Importerer nødvendige funksjoner.

from babel.dates import format_datetime
from datetime import datetime, timedelta
from sympy import Eq, Interval, Rational, solveset, Symbol

# Justerer grafikkingstillingene.

plt.rcParams['axes.grid'] = True
plt.rcParams['figure.figsize'] = 9, 4
```

a) Modell

Løser oppgaven ved å lage en funksjon fra informasjonen i oppgaveteksten.

x er antall dager etter 31. desember 2023.
T(x) gjentar seg hvert år, avhengig av om vi er i et skuddår (2024), ikke-skuddår, eller gjennomsnittsår betyr dette at perioden, p blir 366, 365 eller 365,25 dager. Frekvensen er 2*π/p. Regner ut denne:

```python
periodeforslag = np.array([366, 365.25, 365])
frekvenser = 2/periodeforslag
print(frekvenser)
```

```python
# Alle tre verdier rundes av til tallet i oppgaven.

print(frekvenser.round(4))
```

```python
# Vi bruker skuddåret, siden det helt sikkert kommer en oppdatert versjon for 2025.

periode = 366
frekvens = 2*sp.pi/periode
print(frekvens)
```

```python
# Siden den minste verdien er 15 og den største verdien er 23 blir amplituden og likevektslinjen.

Tmin, Tmax = 15, 23
likevektslinje = Rational(Tmax + Tmin, 2)
print(likevektslinje)
```

```python
amplitude = Rational(Tmax - Tmin, 2)
print(amplitude)
```

Vi ser at vi får likevektsinjen og amplituden fra oppgaven.

Faseforskyvningen er valgt slik at lyset slår seg på 1. april klokken 19. Dette gir oss en likning for faseforskyvningen: f(xa) = Ta.

Dagnummeret hvor vi krysser likevektslinjen er:

```python
xa = 31 + 29 + 31 + 1 + Rational(1, 2)
```

De tre første leddene er antall dager i januar, februar og mars. siden dagnummeret starter på 1 blir det neste leddet 1. Det siste leddet gjør at vi krysser likevektslinjen midt på dagen. 12:00.

Vi kan nå sette opp likningen.

```python
Ta = 19
d = Symbol("d", positive=True)
likning = Eq(amplitude*sp.sin(frekvens*(xa - d)) + likevektslinje, Ta)
```

Løser likningen med solveset.

solveset finner bare løsninger på trigonometriske løsninger dersom alle tallverdier er eksakte (IKKE desimaltall). Dersom noen av tallverdiene er desimaltall må vi bruke Sympy-funksjonen solve eller numeriske metoder til å finne løsningene.

```python
løsning = solveset(likning, d, Interval(0, 366))
print(løsning)
```

```python
# Gjør løsningen om til tall.

løsninger = list(løsning.evalf())
print(løsninger)
```

Den ene løsningen tilsvarer at tidspunktet blir senerer 1. april, den andre løsningen tilsvarer at tiden blir kortere. Dersom vi vil at tidspunktet skal bli senere (noe som er rimelig, siden det blir senere mørkt) bruker vi det minste svaret. Konverterer svaret til brøk med Rational.

```python
dv = Rational(min(løsninger))
```

```python
# Dette gir oss et uttrykk for T(x).

x = Symbol("x")
T = amplitude*sp.sin(frekvens*(x - dv)) + likevektslinje
print(f"3a) T(x) = {T}")
```

```python
# Funksjonen blir tilnærmet lik funksjonen i oppgaven. Vi kan tegne denne.

sp.plot(T, (x, 0, 366))
```

Oppgavetekst

De fem første kubikktallene er 13, 23, 33, 43 og 53. Se figurene ovenfor. La Sn være summen av de n første kubikktallene.

a) Beskriv den rekursive sammenhengen mellom Sn og Sn+1. Bestem en eksplisitt formel for Sn.

b) Lag et program som regner ut S50 ved å bruke den rekursive sammenhengen du fant i oppgave a.

c) Bruk induksjonsbevis til å bevise den eksplisitte formelen for Sn.

```python
# Importerer nødvendige biblioteker

import numpy as np

# Importerer nødvendige funksjoner

from sympy import Eq, Function, linsolve, rsolve, Symbol, symbols
```

Uttrykk for S - Rekursivt uttrykk

Den rekursive sammenhengen blir:

S_1 = 1^3

S_2 = S_1 + 2^3

S_3 = S_2 + 3^3

S_n = n^3 + S_{n+1}

Eksplisitt uttrykk

Løsning med Sympy-funksjonen rsolve

Bruker Sympy-funksjonen rsolve til å finne et eksplisitt uttrykk fra rekursjonen.

```python
n = Symbol("n")
S = Function('S')
f = S(n) - n**3 - S(n-1)
eksplisitt = rsolve(f, S(n), {S(1): 1})
print(f'4a) Eksplisitt uttrykk: S(n) = {eksplisitt} = {eksplisitt.expand()}')
```

```python
# Tester

print([eksplisitt.subs(n, i) for i in range(1, 6)])
```

Vi ser at svaret stemmer med oppgaven.

Løsning med likningssett
Vi ser at det er et n^3-ledd i det rekursive uttrykket. Dette gjør at vi misteker at det eksplisitte uttrykket er et fjerdegradsuttrykk. Lager et likningssett for å finne koeffisientene i fjerdegradsuttrykket. Vi bruker ikke regresjon fordi løsningen skal passe eksakt.

Regner ut den første kubetallene.

```python
N = np.arange(1, 6)
kubetall = N**3
print(kubetall)
```

```python
# Regner ut delsummene

delsummer = kubetall.cumsum()
print(delsummer)
```

```python
# Lager en generisk fjerdegradsfunksjon.

a, b, c, d, e= symbols(["a","b","c","d","e"])
f = a*n**4 + b*n**3 + c*n**2 + d*n + e
koeffisienter = [a, b, c, d, e]

# Lager et likningssett vi kan bruke til å finne koeffisientene.

likningssett = [Eq(f.subs(n, i), s) for i, s in zip(N, delsummer)]

# Løser likningssettet med linsolve.

løsning = linsolve(likningssett, koeffisienter)
print(løsning)
```

```python
# Setter koeffisientene inn i det generiske uttrykket og får et eksplisitt uttrykk for Sb.

likning = f.subs(dict(zip(koeffisienter, list(løsning)[0])))
print(f'4a) Eksplisitt uttrykk: S(n) = {likning}')
```

```python
# Oppsett
# Importerer nødvendige biblioteker.

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Importerer nødvendige funksjoner.

from sympy import lambdify, Plane, Point, symbols, solveset
```

```python
# Løsning med Sympy-geometri
# a) Minste avstand fra kuleflate til plan
# Legger inn punktene på kuleflaten som Point-objekter.

A = Point(1, 2, 1)
B = Point(3, 0, -3)

# Finner sentrum, S til kulen.

S = (A + B)/2
print(S)
```

```python
# Finner radien til kulen, r. Dette er halvparten av lengden til avstandsvektoren AB.

R = (A - B)/2
r = sp.sqrt(R.dot(R))
print(r)
```

```python
# Legger inn planet γ: x + 2y + 2x = 14 som et Plane-objekt. Normalvektoren er koeffisientene foran x-, y og z-leddene. 
# Løser likningen for x når y = 0 og z=0 for å finne ut at (14, 0, 0) er et vilkårlig punkt i planet.

normalvektor = np.array([1, 2, 2])
gamma = Plane(Point(14, 0, 0), normal_vector=normalvektor)

# Finner avstanden fra plan til sentrum med Plane-metoden distance.

print(gamma.distance(S))
```

```python
# Trekker fra radien for å finne den minste avstanden fra kuleflaten til planet.

avstand = gamma.distance(S) - r
print(f'5a) Den minste avstanden mellom kuleflaten og planet er {avstand} = {avstand.evalf(3)}')
```

```python
# b) Likning for planet α
# Planet α er på den andre siden av kulesenteret, S.

# Finner først den perpendikulære linjen til planet gjennom kulesenteret med metoden perpendicular_line. Dette er et objekt av typen Line,

l = gamma.perpendicular_line(S)
print(l)
```

```python
# Finner skjæringspunkt mellom linje og plan med metoden intersection.

P = Point(l.intersection(gamma)[0])
print(P)
```

```python
# Kan finne et vilkårlig punkt, T på α med vektorregning.
# T = P + 2SP

T = P + 2*(S - P)
print(T)
```

```python
# Vi kan nå finne det parallelle planet til ɣ som T ligger på med metoden paralell_plane.

alpha = gamma.parallel_plane(T)
print(alpha)
```

```python
# Vi henter ut likningen til planet med metoden equation.

print(f'5b) α: {alpha.equation()}')
```

```python
# Vi kan teste om avstanden til kuleflaten er den samme for de to planene.

print(gamma.distance(S) - r, "==", alpha.distance(S) - r)
```

```python
print(gamma.distance(S) - r == alpha.distance(S) - r)
```

Svaret er rimelig.

# Eksamen R2 vår 2024: Oppgave 5 - Del 2 Grafisk fremstilling

Grafisk framstilling
Vi kan tegne kulen sammen med de to planene.

Kule
Lager først en funksjon for å tegne en kule.

```python
def tegn_kule(sentrum, radius, fig = None):
    #Lager et figurobjekt dersom det ikke finnes.
    if fig:
        ax = fig.gca()
    else:
        fig = plt.figure()
        ax = fig.add_subplot(projection='3d')
 
    #Bruker en parameterframstilling for overflaten.
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    kule_x = radius * np.outer(np.cos(u), np.sin(v)) + sentrum[0]
    kule_y = radius * np.outer(np.sin(u), np.sin(v)) + sentrum[1]
    kule_z = radius * np.outer(np.ones(np.size(u)), np.cos(v)) + sentrum[2]
 
    #Tegner overflaten
    ax.plot_surface(kule_x, kule_y, kule_z, color="red", alpha=.4)
 
    #Justerer grafvinduet
    ax.axis('scaled')
    fig.tight_layout()
    fig.set_size_inches(7, 7 )
    return fig, ax

# Tegner kulen sammen med punktene A, B, diameteren AB og kulesenteret.

sentrum = [float(e) for e in S]
radius = float(r)
fig, ax = tegn_kule(sentrum, radius)
ax.scatter(*A)
ax.text(*A, "A")
ax.scatter(*B)
ax.text(*B, "B")
ax.scatter(*S)
ax.text(*S, "S")
ax.plot(*zip(A, B), color="black", linestyle="--")
plt.show()
```

Vi ser at AB er en diameter. Vi fant den riktige kulen.

```python
# Plan
# Vi lager så en funksjon for å tegne et plan.

def tegn_plan(plan, sentrum=[0, 0, 0], radius = 1, fig=None):
    #Lager en figur og akse dersom de ikke finnes fra før
    if fig:
        ax = fig.gca()
    else:
        fig = plt.figure()
        ax = fig.add_subplot(projection='3d')
 
    #Finner x- og y- verdiene vi skal tegne planet for, sentrert rundt et gitt punkt.
    xlim = sentrum[0] + radius*np.array([-1, 1])
    ylim = sentrum[1] + radius*np.array([-1, 1])
    zlim = sentrum[2] + radius*np.array([-1, 1])
    ax.scatter(*sentrum)
 
    #Lager et x,y-rutenett.
    X = np.linspace(*xlim, 2)
    Y = np.linspace(*ylim, 2)
    X, Y = np.meshgrid(X, Y)
 
    #Løser likningen for planet for å få et uttrykk for z
    x, y, z = symbols(["x", "y", "z"])
    z_uttrykk = list(solveset(plan.equation(x, y, z), z))[0]
 
    #Bruker funksjonen [sy.Lambdify] til å lage en funksjon for z.
    fun = lambdify((x, y), z_uttrykk)
    Z = fun(X, Y)
 
    #Tegner planet
    ax.plot_surface(X, Y, Z, color="grey", alpha=0.3)
 
    #Vi setter grensene for grafvinduet
    ax.set_xlim(*xlim)
    ax.set_ylim(*ylim)
    ax.set_zlim(*zlim)
    fig.tight_layout()
    fig.set_size_inches(7, 7)
    return fig, ax

# Konverterer de nærmeste punktene til planene fra Sympy-objekter til tall. Vi sentrerer planene rundt disse to punktene.

Pn = [float(e) for e in P]
Tn = [float(e) for e in T]

#Tegner kulen og de to planene. Legger til kulesenteret og de nærmeste punktene på de to planene.

fig, ax = tegn_kule(sentrum, radius)
fig, ax = tegn_plan(gamma, sentrum = Pn, radius=5, fig=fig)
fig, ax = tegn_plan(alpha, sentrum = Tn, radius=5, fig=fig)
ax.scatter(*S)
ax.text(*S, "S")
ax.scatter(*P)
ax.text(*P, "P")
ax.scatter(*T)
ax.text(*T, "T")
ax.plot(*zip(P, T), color="black")
ax.set_xlim(sentrum[0] - 4, sentrum[0] + 4)
ax.set_ylim(sentrum[1] - 4, sentrum[1] + 4)
ax.set_zlim(sentrum[2] - 4, sentrum[2] + 4)
plt.show()
```

Løsningen vår ser rimelig ut.

# Eksamen R2 vår 2024: Oppgave 6 - Del 2



Denne oppgaven fra del 2 handler om å summere en geometrisk rekke hvor kvotienten er verdien av et integral. Vi skal løse oppgaven med Sympy. Oppgaven ber oss om å finne verdien på en ukjent konstant. Avhengig av antakelsene vi gjør blir verdien på denne konstanten 1, 2 eller udefinert. Det at en oppgave krever antakelser og kan ha flere løsninger er en trend vi har sett på de siste årenes eksamensett.

Oppgavetekst
La a₁ > 0 og la S(x) være summen av en rekke gitt ved

$latex S(x) = \Sum_{n=0}^{\infty} a_1 \cdot (\Int_0^x e^{-t} dt)^n

Bestem a₁ slik at den minste mulige summen blir 1.

```python
# Oppsett
# Importerer nødvendige biblioteker.

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Importerer nødvendige funksjoner.

from sympy import limit, Interval, minimum, oo, Reals, solveset
from sympy import summation, Symbol

# Justerer grafikkingstillingene.

plt.rcParams['axes.grid'] = True
plt.rcParams['figure.figsize'] = 9, 4
```

Løsning for x større enn 0
Innledende betraktinger:

Antar fra oppgaven at x > 0. Dette er fordi elever på videregående skole bare blir presentert for situasjoner hvor øvre integrasjonsgrense er større enn nedre integrasjonsgrense. Jeg skal senere i oppgaven se vekk i fra denne begrensningen.
Ser at S er summen av en geometrisk rekke siden hvert ledd er et multiplum av det forrige. Forholdet mellom etterfølgende ledd er verdien av integralet.
Trekker a₁ ut av summen og finner et uttrykk for denne summen. Regner først ut integralet med integrate-metoden.

```python
t = Symbol("t", real=True)
x = Symbol("x", positive=True, real=True)
integral = sp.exp(-t).integrate((t, 0, x))
print(integral)
```

```python
# Skisserer verdiene.

sp.plot(integral, (x, -2, 4))
```

```python
# Regner så ut summen med summation-funksjonen.

n = Symbol("n")
S = summation(integral**n, (n, 0, oo))
print(S)
```

```python
# Forsøker å forenkle utttrykket.

S = S.simplify()
print(S)
```

```python
# Finner den minste verdien til summen for x > 0.

Smin = minimum(S, x, Interval(0, oo, left_open =True))
print(Smin)
```

```python
# Det ekstra argumentet i Interval-konstruktøren gjør at x = 0 ikke blir en del av intervallet. (Åpent intervall)
# Den minste verdien a må ha for at summen alltid skal være større enn eller lik 1 er.

a1 = 1/Smin
print(f"6) Den minste verdien a_1 kan ha er {a1}")
```
